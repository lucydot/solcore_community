## Mapping out the Solcore community

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from itertools import chain
```

```python
# create basemap
fig = plt.figure(figsize=(10, 4),constrained_layout=True)
gs = fig.add_gridspec(3, 4)
ax1 = fig.add_subplot(gs[:, 0:3])

m = Basemap(projection='cyl', resolution=None,
            llcrnrlat=-60, urcrnrlat=80,
            llcrnrlon=-130, urcrnrlon=160)
m.shadedrelief(scale=0.2,alpha=0.6)

# latitudes and longitudes of cities where Solcore is used / developed.

#  cont = Github contributor
lat_cont = [51.509865,-8.0577,45,-33,12,55.45,35,52,45,45,-37,51]
lon_cont = [-0.118092,-35.8830,-80,151,77,37.37,51,13,12,9,144,1]
# london, recife, toronto, bangalore, sydney, moscow, iran, berlin, venice, italy, reading

# cite = Google citation (Aug 21)
lat_cite = [ 33, 41.46, 35.11,42, 40,59,39,45,53,49,23,52]
lon_cite =[ -7,-72.4,136,-71, -105,17,-75,-75,-2,6,90,-1]
# morocco, hartford, nagoya, boston, colorado,uppsala, ottowa,manchester,metz,dhaka,loughborough

#  fork = Github forks
lat_fork = [22,-12,48.8566,52,28,35,26,61,34]
lon_fork = [-100,-72,2.3522,-1,77,139,80,23,-118]
# mexico(potosi), lima,paris, birmingham, new delhi, tokyo, kanpur,tampere, LA

# Plot  cities on  map
m.scatter(lon_cont,lat_cont,marker='o',color='darkorange',latlon=True,s=15,label="Developer")
m.scatter(lon_cite,lat_cite,marker='o',color='darkblue',latlon=True,s=15,label="User")# london
m.scatter(lon_fork,lat_fork,marker='o',color='darkblue',latlon=True,s=15)# london
#m.scatter(-34.8830,-8.0577,marker='.',latlon=True)# recife

plt.legend(loc=3)
plt.title("Solcore5 community map")


ax2 = fig.add_subplot(gs[0, 3])
ax2.bar([0.5,1.5,2.5,3.5],[3,15,5,2],width=0.9)
ax2.set_xticks([0.5,1.5,2.5,3.5])
ax2.set_xticklabels(['2018','2019','2020','2021'])
ax2.set_yticks([5,10,15])
ax2.tick_params(axis='x',size=0)
ax2.set_title("# Unique code contributors")
ax3 = fig.add_subplot(gs[1, 3])
ax3.bar([0.5,1.5,2.5,3.5],[4,10,15,10],width=0.9)
ax3.set_xticks([0.5,1.5,2.5,3.5])
ax3.set_xticklabels(['2018','2019','2020','2021'])
ax3.set_yticks([5,10,15])
ax3.tick_params(axis='x',size=0)
ax3.set_title("# Citations")
ax4 = fig.add_subplot(gs[2, 3])
#ax4.set_title("# Github forks")
#ax4.bar([0.5,1.5,2.5,3.5],[0,11,7,12],width=0.9)
ax4.set_title("# PyPI downloads")
ax4.bar([0.5,1.5,2.5,3.5],[1697, 2661, 4073, 3613],width=0.9)
ax4.set_xticks([0.5,1.5,2.5,3.5])
ax4.set_xticklabels(['2018','2019','2020','2021'])
ax4.set_yticks([1000,2000,3000,4000])
ax3.tick_params(axis='x',size=0)

11,7,12


plt.savefig('Solcore_community.png')
```

Note: 
- The latest iteration of Solcore, Solcore5, was released at the end of 2017.
- Developer corresponds to someone who is a contributor to the Solcore 5 repo on Github. 
- User(citation) corresponds to a citation on Google scholar (as of Aug 2021)
- User(Github fork) corresponds to someone who has forked Solcore on Github
- This does not reflect the full user and developer base as some locations are not given on Github profiles
- In many cases a developer will also be a user as defined here, but  this  is not reflected in the map above.
- The full world map  is not shown....Solcore isn't used  in the  North or  South pole but I'm sure we can work on that :)
